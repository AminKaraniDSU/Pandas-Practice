## Drope Rows and Columns

```python
import pandas as pd
```

## Filters Rows

```python
movie = pd.read_csv('https://bit.ly/imdbratings')
```

```python
movie[movie.duration>200]
```

```python
movie.loc[movie.duration>200,'genre']
```

```python
fil_list = ['Drama','Crime']
movie.loc[(movie.duration>200) & (movie.genre.isin(fil_list))]
```
